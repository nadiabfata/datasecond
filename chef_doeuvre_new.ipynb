## II- Quels sont les pays les plus chauds du monde ?

```python
city=drop_na.groupby(['City','Country'])['AverageTemperature'].mean().reset_index().sort_values(by='AverageTemperature',ascending=False)
city=city.drop_duplicates(subset='Country',keep='first')
city=city.set_index(['City','Country'])
plt.subplots(figsize=(4,10))
sns.barplot(y=city.index,x='AverageTemperature',data=city,palette='RdYlGn').set_title('Villes les plus chaudes par pays')
plt.xlabel('Température moyenne')
```

Le barplot expose trois niveaux différents du réchauffement climatique à l’échelle globale. Pour illustrer, les pays comme Sudan, India Saudi Arabie sont classifiés parmi les pays les plus chaudes au monde tandis que Russia, Chile et Canada sont classés parmi les payer les plus froid.

### 2- Est-ce que le CO2 est responsable du réchauffement climatique?

```python
temp_df = drop_first_row2[drop_first_row2['Entity']== 'France'   ]
temp_df['year'] = pd.to_datetime(temp_df['Year']).dt.year

by_year = temp_df.groupby('Year')['CO2'].mean().reset_index()
by_year.iplot(kind='scatter', x='Year', y='CO2', title='CO2 france',
             xTitle='Years', yTitle='CO2', legend=True)
```

# Concentration d’ozone (O3)
https://ourworldindata.org/grapher/ozone-o3-concentration-in-ppb

```python
ozone = pd.read_csv(r'C:\Users\33788\Desktop\dossier_data_all\chef_doeuvre\research_paper\ozone-o3-concentration-in-ppb.csv')
ozone.head()
```

```python
ozone.rename(columns={'Ozone concentration (parts per billion)': 'ozone', }, inplace=True)
ozone.head()
```

```python
pd.set_option('display.max_columns', 500)
```

```python
data = dict(type="choropleth",
           locations = ozone['Entity'],
            locationmode = "country names",
           z = ozone['ozone'],
           text = ozone['ozone'],
           colorbar = {'title':'Ozone(O3)'})

layout = dict(title="Ozone(O3): Concentration",
             geo = dict(showframe=False,
                      projection = {'type':'mercator'}))
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

Concentrations moyennes pondérées en fonction de la population d'ozone (O₃) en parties par milliard (ppb). Les concentrations locales d'ozone sont enregistrées et estimées à une résolution de 11x11 km. Ces valeurs sont ensuite pondérées par la
densité de population pour le calcul des concentrations moyennes au niveau national.
